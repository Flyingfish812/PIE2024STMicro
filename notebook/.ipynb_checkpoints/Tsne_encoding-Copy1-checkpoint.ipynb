```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Compute differences between paired numerical features
feature_pairs = [
    ('Maximum Input Offset Voltage', 'Maximum Input Offset Voltage_comp', 'd_offset_voltage'),
    ('Maximum Single Supply Voltage', 'Maximum Single Supply Voltage_comp', 'd_max_supply_voltage'),
    ('Minimum Single Supply Voltage', 'Minimum Single Supply Voltage_comp', 'd_min_supply_voltage'),
    ('Number of Channels per Chip', 'Number of Channels per Chip_comp', 'd_channels'),
    ('Typical Gain Bandwidth Product', 'Typical Gain Bandwidth Product_comp', 'd_gain_bandwidth')
]

for feat1, feat2, diff_name in feature_pairs:
    df[diff_name] = df[feat1] - df[feat2]

diff_columns = [pair[2] for pair in feature_pairs]

# Step 2: Compute the "distance" between features
# Compute the absolute differences
for diff_name in diff_columns:
    df[f'abs_{diff_name}'] = df[diff_name].abs()

abs_diff_columns = [f'abs_{diff_name}' for diff_name in diff_columns]

# Compute the sum of absolute differences
df['sum_abs_differences'] = df[abs_diff_columns].sum(axis=1)

# Normalize the distance to [0, 1]
# Use the 95th percentile of the sum_abs_differences as a proxy for the maximum distance
max_distance = df['sum_abs_differences'].quantile(0.95)
if max_distance == 0:  # Avoid division by zero
    max_distance = 1.0

df['distance'] = df['sum_abs_differences'] / max_distance
df['distance'] = df['distance'].clip(lower=0.0, upper=1.0)  # Ensure distance is in [0, 1]

# Step 3: Define base scores for each Cross Reference Type
base_scores = {
    'A': 0.85,  # High score for class A (most similar)
    'B': 0.65,
    'C': 0.45,
    'D': 0.25   # Low score for class D (least similar)
}

# Map the base scores to each row
df['base_score'] = df['Cross Reference Type'].map(base_scores)

# Step 4: Compute Y deterministically by adjusting the base score using the distance
spread = 0.15  # Controls how much the distance affects the score
df['Y'] = df['base_score'] - spread * df['distance']

# Clip Y to ensure it stays within [0, 1]
#df['Y'] = df['Y'].clip(lower=0.0, upper=1.0)

# Step 5: Visualize the distributions with multiple methods

# Create a figure with subplots
plt.figure(figsize=(15, 10))

# Visualization 1: Histogram
plt.subplot(2, 2, 1)
sns.histplot(data=df, x='Y', hue='Cross Reference Type', multiple='stack', bins=30)
plt.title('Histogram of Y by Cross Reference Type')
plt.xlabel('Y')
plt.ylabel('Count')

# Visualization 2: Box Plot
plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='Cross Reference Type', y='Y')
plt.title('Box Plot of Y by Cross Reference Type')
plt.xlabel('Cross Reference Type')
plt.ylabel('Y')

# Visualization 3: Strip Plot
plt.subplot(2, 2, 3)
sns.stripplot(data=df, x='Cross Reference Type', y='Y', hue='Cross Reference Type', jitter=True, alpha=0.6)
plt.title('Strip Plot of Y by Cross Reference Type')
plt.xlabel('Cross Reference Type')
plt.ylabel('Y')

# Visualization 4: KDE Plot with Reduced Bandwidth
plt.subplot(2, 2, 4)
sns.kdeplot(data=df, x='Y', hue='Cross Reference Type', fill=True, common_norm=False, bw_adjust=0.1)
plt.title('KDE Plot of Y by Cross Reference Type (Tight Bandwidth)')
plt.xlabel('Y')
plt.ylabel('Density')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

# Display the updated dataframe
print(df[['Cross Reference Type', 'Y', 'distance', 'sum_abs_differences'] + diff_columns].head())
```
