```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
x= np.linspace(0,1.2,1000)
yA=norm.rvs(size=10000, loc=0.96, scale=0.0107)
yB=norm.rvs(size=10000, loc=0.85, scale=0.0249)
yC = norm.rvs(size=10000, loc=0.72, scale=0.01779)
plt.hist(yA, bins=30, density=True,alpha=0.75)
plt.hist(yB, bins=30, density=True,alpha=0.75)
plt.hist(yC, bins=30, density=True,alpha=0.75)
plt.legend(['A','B','C'])
```
