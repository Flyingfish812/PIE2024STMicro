Désequilibre des classes est un probleme car un model simple va prédire la class la plus fréquente donc on risque davoir des scores plus proches de C D et SF: 
[Article sur comment resoudre ce problem](https://www.analyticsvidhya.com/blog/2020/07/10-techniques-to-deal-with-class-imbalance-in-machine-learning/#:~:text=Non%2Dfraudulent)
- Undersampling (cause a loss of information)
- Oversampling (can cause overfishing.)

Intuitivement je pense que undersampling la classe C et B puis oversampling les autres classe est un bon contrepromis  entre les deux.

```python
class_A = df_filtered[df_filtered["Cross Reference Type"] == "A"]
class_B = df_filtered[df_filtered["Cross Reference Type"].isin(["B","B/Upgrade","B/Downgrade"])]
class_C = df_filtered[df_filtered["Cross Reference Type"].isin(["C","C/Upgrade","C/Downgrade"])]
# class_Bup = df_filtered[df_filtered["Cross Reference Type"] == "B/Upgrade"]
# class_Bdown = df_filtered[df_filtered["Crossm Reference Type"] == "B/Downgrade"]
# class_Cup = df_filtered[df_filtered["Cross Reference Type"] == "C/Upgrade"]
# class_Cdown = df_filtered[df_filtered["Cross Reference Type"] == "C/Downgrade"]
n=20000
random_state = 42
```

```python
class_B_downsampled = class_B.sample(n=n, random_state=random_state)
class_C_downsampled = class_C.sample(n=n,random_state=random_state)
class_A_oversample = class_A.sample(n=n,random_state=random_state, replace= True)
# class_Bup_oversample = class_Bup.sample(n=n,random_state=random_state, replace= True)
# class_Bdown_oversample = class_Bdown.sample(n=n,random_state=random_state, replace= True)
# class_Cup_oversample = class_Cup.sample(n=n,random_state=random_state, replace= True)
# class_Cdown_oversample = class_Cdown.sample(n=n,random_state=random_state, replace= True)
```

```python
df_sampled = pd.concat([class_A_oversample, class_B_downsampled, class_C_downsampled],axis=  0)
df_sampled["Cross Reference Type"].value_counts()
```

```python
sns.countplot(df_sampled["Cross Reference Type"])
```

```python
from scipy.stats import norm
base_means = {
    'A': 0.96,
    'B': 0.85,
    'C': 0.72
}
base_std={
    'A': 0.0107,
    'B': 0.0249,
    'C': 0.01779
}
def get_params(category):
    parts = category.split('/')
    mean = base_means.get(parts[0], 0.0)
    std = base_std.get(parts[0], 0.0)
    for part in parts[1:]:
        if part == 'Upgrade':
            mean += 0.04
        elif part == 'Downgrade':
            mean -= 0.04
    return mean, std
df_sampled["Mean"] , df_sampled["Std"] = zip(*df_sampled["Cross Reference Type"].map(get_params))
df_sampled['Closeness'] = df_sampled.apply(lambda row: norm.rvs(loc=row['Mean'], scale=row['Std']), axis=1)
df_sampled['Closeness'] = df_sampled['Closeness'].clip(0, 1)

```

Gower's distance
Pour deux objets \( i \) et \( j \) ayant \( p \) descripteurs, la similarité \( S \) est définie comme :

$$
S_{ij} = \frac{\sum_{k=1}^{p} w_{ijk} s_{ijk}}{\sum_{k=1}^{p} w_{ijk}},
$$

où les poids \( w_{ijk} \) sont des coefficients non négatifs généralement fixés à 1, et \( s_{ijk} \) est la similarité entre les deux objets concernant leur \( k \)-ème variable.

- **Variables Binaires ou Ordinales :**

  $$
  s_{ijk} =
  \begin{cases}
    1 & \text{si égaux}, \\
    0 & \text{sinon}.
  \end{cases}
  $$

- **Variables Continues :**

  $$
  s_{ijk} = 1 - \frac{|x_i - x_j|}{R_k},
  $$

  où $ R_k $ est l'étendue de la \( k \)-ème variable, assurant ainsi que $0 \leq s_{ijk} \leq 1 $.

Par conséquent, la similarité globale $ S_{ij} $ entre deux objets est la moyenne pondérée des similarités calculées pour tous leurs descripteurs.

Dans son exposition originale, la distance ne traite pas les variables ordinales de manière particulière. Dans les années 1990, d'abord Kaufman et Rousseeuw, puis Podani, ont suggéré des extensions où l'ordre d'une caractéristique ordinale est utilisé. Par exemple, Podani obtient des différences de rang relatives comme suit :

$$
s_{ijk} = 1 - \frac{|r_i - r_j|}{\max\{r\} - \min\{r\}},
$$

où \( r \) représente les rangs correspondant aux catégories ordonnées de la \( k \)-ème variable.

```python
categorical_descriptors = ['MANUFACTURER', 'Supplier_Package']
numerical_descriptors = [
    'Maximum Input Offset Voltage',
    'Maximum Single Supply Voltage',
    'Minimum Single Supply Voltage',
    'Number of Channels per Chip',
    'Typical Gain Bandwidth Product'
]

# Calculate similarity for categorical descriptors
for desc in categorical_descriptors:
    comp_desc = f"{desc}_comp"
    similarity_col = f"{desc}_similarity"
    df_sampled[similarity_col] = (df_sampled[desc] == df_sampled[comp_desc]).astype(int)

# Calculate similarity for numerical descriptors
for desc in numerical_descriptors:
    comp_desc = f"{desc}_comp"
    similarity_col = f"{desc}_similarity"
    
    # Calculate the range for normalization
    combined = pd.concat([df_sampled[desc], df_sampled[comp_desc]])
    desc_range = combined.max() - combined.min()
    
    # Avoid division by zero
    if desc_range == 0:
        df_sampled[similarity_col] = 1
    else:
        df_sampled[similarity_col] = 1 - (df_sampled[desc] - df_sampled[comp_desc]).abs() / desc_range
        df_sampled[similarity_col] = df_sampled[similarity_col].clip(0, 1) # Clip to [0, 1]

# Display the first few rows to verify
df_sampled[[i+"_similarity" for i in numerical_descriptors]].describe()
#df_sampled[df_sampled["Cross Reference Type"]=="B"].iloc[10000:10015]
df_sampled.columns
```
